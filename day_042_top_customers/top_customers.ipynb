```python
import pandas as pd

df = pd.read_csv("ecommerce_cleaned.csv")
df.head()
```

```python
customer_spend = df.groupby("customer_id")["total_amount"].sum()
customer_spend
```

```python
customer_spend = customer_spend.sort_values(ascending=False)
customer_spend
```

```python
top_10_customers = customer_spend.head(10)
top_10_customers
```

```python
total_revenue = df["total_amount"].sum()
top_10_revenue = top_10_customers.sum()

percentage = (top_10_revenue / total_revenue) * 100
percentage
```

Calculated total spending per customer.
Identified top 10 customers by revenue.
Checked what percentage of total revenue they contribute.
