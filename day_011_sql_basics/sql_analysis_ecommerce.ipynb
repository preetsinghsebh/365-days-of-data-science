# Day 11 – SQL-Style Business Analysis
## E-Commerce Customer Behavior & Sales Analysis

**Objective:**  
Answer key business questions using SQL-style logic (GROUP BY, WHERE, ORDER BY, HAVING)
implemented with pandas.

```python
import pandas as pd

df = pd.read_csv("/content/ecommerce_cleaned.csv")
df.head()
```

```python
df["total_amount"].sum()
```

```python
df.groupby("product_category")["total_amount"].sum()
```

```python
df.groupby("product_category")["total_amount"].sum().sort_values(ascending=False)
```

```python
df.groupby("product_category")["order_id"].count()
```

```python
df.groupby("customer_id")["total_amount"].sum()
```

```python
df.groupby("customer_id")["total_amount"].sum().sort_values(ascending=False).head(5)
```

```python
if "payment_method" in df.columns:
    df.groupby("payment_method")["order_id"].count()
```

## Key Observations

- Some product categories generate more revenue than others.
- A small number of customers contribute higher total revenue.
- Customer spending is not evenly distributed.
- Certain payment methods are used more frequently.
