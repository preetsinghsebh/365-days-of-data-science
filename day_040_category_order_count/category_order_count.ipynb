```python
import pandas as pd

df = pd.read_csv("ecommerce_cleaned.csv")
df.head()
```

```python
category_orders = df.groupby("product_category")["order_id"].count()
category_orders
```

```python
category_orders = category_orders.sort_values(ascending=False)
category_orders
```

```python
category_revenue = df.groupby("product_category")["total_amount"].sum()

comparison = pd.DataFrame({
    "total_orders": category_orders,
    "total_revenue": category_revenue
})

comparison.sort_values("total_orders", ascending=False)
```

Calculated total number of orders per product category.
Compared order volume with revenue.
