## Objective
Explore customer behavior and sales trends using simple aggregations and plots.

## Dataset
E-Commerce Customer Behavior & Sales Analysis (Kaggle)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/ecommerce_cleaned.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df["total_amount"].sum()
```

```python
df.groupby("product_category")["total_amount"].sum().sort_values(ascending=False)
```

```python
df["product_category"].value_counts()
```

```python
df.groupby("product_category")["total_amount"].sum().plot(kind="bar")
plt.title("Total Sales by Product Category")
plt.show()
```

```python
df["payment_method"].value_counts().plot(kind="bar")
plt.title("Payment Method Distribution")
plt.show()
```

## Key Observations

- Some product categories generate more revenue than others.
- Electronics and Home & Garden are among the top revenue contributors.
- Certain payment methods are used more frequently by customers.

This analysis provides a basic understanding of customer purchasing behavior.
