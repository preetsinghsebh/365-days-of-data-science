# E-Commerce Revenue by Product Category
## Focused Analysis

This notebook analyzes how revenue is distributed across different
product categories in an e-commerce dataset.

The goal is to identify top-performing categories and extract
clear business insights.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/ecommerce_eda.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
category_revenue = (
    df.groupby("product_category")["total_amount"]
      .sum()
      .sort_values(ascending=False)
)

category_revenue
```

```python
plt.figure()
category_revenue.plot(kind="bar")
plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.show()
```

## Key Insights

- A small number of product categories contribute most of the revenue.
- Some categories have high order volume but lower revenue contribution.
- Focusing on top categories can improve marketing and inventory decisions.

This focused analysis highlights how category-level insights
can support business strategy.
