# E-Commerce Category Revenue Insights
## Focused Business Analysis

This notebook focuses on understanding how revenue is distributed
across product categories in an e-commerce business.

The objective is not broad exploration, but identifying
which categories drive the majority of revenue and why that matters.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/ecommerce_eda.csv")
df.head()
```

```python
df[["product_category", "total_amount"]].head()
```

```python
category_revenue = (
    df.groupby("product_category")["total_amount"]
      .sum()
      .sort_values(ascending=False)
)

category_revenue
```

```python
plt.figure()
category_revenue.plot(kind="bar")
plt.title("Revenue Contribution by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.show()
```

## Interpretation & Business Impact

- Revenue is highly concentrated in a small number of categories.
- Not all high-order-volume categories generate high revenue.
- Focusing marketing and inventory decisions on top categories
  can significantly impact overall business performance.

This focused view helps decision-makers prioritize actions
instead of treating all categories equally.
