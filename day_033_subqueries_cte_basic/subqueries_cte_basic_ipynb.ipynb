```python
import pandas as pd
import sqlite3

df = pd.read_csv("ecommerce_cleaned.csv")
conn = sqlite3.connect(":memory:")
df.to_sql("orders", conn, index=False, if_exists="replace")
```

```python
query = """
SELECT customer_id,
       SUM(total_amount) AS total_spent
FROM orders
GROUP BY customer_id
HAVING total_spent >
       (SELECT AVG(total_amount) FROM orders);
"""
pd.read_sql(query, conn)
```

```python
query = """
WITH customer_spend AS (
    SELECT customer_id,
           SUM(total_amount) AS total_spent
    FROM orders
    GROUP BY customer_id
)
SELECT *
FROM customer_spend
WHERE total_spent >
      (SELECT AVG(total_amount) FROM orders);
"""
pd.read_sql(query, conn)
```

### Key Insight
Subqueries and CTEs help break complex problems into simple steps.
They make SQL easier to read and reason about.
