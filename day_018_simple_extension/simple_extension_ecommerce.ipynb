## Category vs Payment Method Analysis

Objective:
Understand how payment preferences vary across top product categories.

```python
import pandas as pd

df = pd.read_csv("/content/ecommerce_customer_behavior_dataset.csv")
df.head()
```

```python
top_categories = (
    df.groupby("Product_Category")["Total_Amount"]
      .sum()
      .sort_values(ascending=False)
      .head(3)
      .index
)

top_categories
```

```python
payment_by_category = (
    df[df["Product_Category"].isin(top_categories)]
      .groupby(["Product_Category", "Payment_Method"])
      .size()
      .reset_index(name="count")
)

payment_by_category
```

```python
payment_by_category.pivot(
    index="Product_Category",
    columns="Payment_Method",
    values="count"
).fillna(0)
```

## Key Insight (Day 18)

- Payment preferences differ across top product categories.
- Some categories rely more on specific payment methods.
- Understanding this helps optimize checkout options and promotions.
