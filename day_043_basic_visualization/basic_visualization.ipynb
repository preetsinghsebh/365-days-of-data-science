```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ecommerce_cleaned.csv")
```

```python
category_revenue = df.groupby("product_category")["total_amount"].sum()
category_revenue = category_revenue.sort_values(ascending=False)

category_revenue
```

```python
plt.figure()
category_revenue.plot(kind="bar")
plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

Created a basic bar chart to visualize revenue by product category.
Visualization makes it easier to compare performance across categories.
