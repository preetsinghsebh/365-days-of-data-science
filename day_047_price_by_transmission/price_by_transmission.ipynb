```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Car details v3.csv")
```

```python
avg_price_transmission = df.groupby("transmission")["selling_price"].mean()
avg_price_transmission = avg_price_transmission.sort_values(ascending=False)

avg_price_transmission
```

```python
avg_price_transmission.plot(kind="bar")
plt.title("Average Selling Price by Transmission Type")
plt.xlabel("Transmission")
plt.ylabel("Average Price")
plt.show()
```

### Insight

Automatic cars generally have higher average selling prices
compared to manual cars.

Transmission type impacts resale value.
