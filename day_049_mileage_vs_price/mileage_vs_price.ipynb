```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Car details v3.csv")
```

```python
df["km_driven"].head()
```

```python
 plt.scatter(df["km_driven"], df["selling_price"])
plt.title("Mileage vs Selling Price")
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling Price")
plt.show()
```

```python
df["km_driven"].corr(df["selling_price"])
```

### Insight

Cars with higher mileage generally have lower selling prices.
Mileage is one of the strongest indicators of vehicle depreciation.
