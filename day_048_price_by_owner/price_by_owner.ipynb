```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Car details v3.csv")
```

```python
df["owner"].value_counts()
```

```python
avg_price_owner = df.groupby("owner")["selling_price"].mean()
avg_price_owner = avg_price_owner.sort_values(ascending=False)

avg_price_owner
```

```python
avg_price_owner.plot(kind="bar")
plt.title("Average Selling Price by Owner Type")
plt.xlabel("Owner Type")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

### Insight

Cars with fewer previous owners tend to have higher selling prices.
Ownership history impacts resale value.
