# Day 15 – Mini Case Study
## E-Commerce Customer Behavior & Sales Analysis

### Problem Statement
An e-commerce business wants to understand its sales performance
and customer behavior to improve revenue and customer retention.

### Dataset
Source: Kaggle – E-Commerce Customer Behavior & Sales Analysis

The dataset contains transactional data including:
- customer identifiers
- product categories
- order values
- payment methods
- purchase history

### Approach

1. Data Cleaning
   - Handled missing values
   - Removed duplicates
   - Standardized column names

2. Exploratory Data Analysis
   - Analyzed revenue trends
   - Evaluated category performance
   - Studied order value distribution

3. Customer Segmentation
   - Grouped customers into High, Medium, and Low value segments
   - Based on spending and purchase frequency

### Key Findings

- A small number of product categories contribute most of the revenue.
- High-value customers represent a small portion of users but drive significant sales.
- Most customers fall into the medium-value segment, offering growth potential.

### Business Recommendations

- Focus marketing efforts on top-performing categories.
- Retain high-value customers using loyalty programs.
- Upsell medium-value customers through targeted offers.
- Re-engage low-value customers with promotions.

### Conclusion

This case study demonstrates how structured data analysis
can translate raw transactional data into actionable business insights.
