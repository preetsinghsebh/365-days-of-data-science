```python
import pandas as pd

df = pd.read_csv("/content/ecommerce_cleaned.csv")
df.head()
```

```python
orders = df[["order_id", "customer_id", "total_amount"]]
orders.head()
```

```python
customers = df[["customer_id", "city"]].drop_duplicates()
customers.head()
```

```python
inner_join = pd.merge(
    orders,
    customers,
    on="customer_id",
    how="inner"
)

inner_join.head()
```

inner_join = pd.merge(

    on="customer_id",
    how="inner"
)

inner_join.head()

```python
left_join = pd.merge(
    orders,
    customers,
    on="customer_id",
    how="left"
)

left_join.head()
```

SELECT *
FROM orders
LEFT JOIN customers
ON orders.customer_id = customers.customer_id;

```python
region_revenue = (
    left_join
    .groupby("city")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

region_revenue
```

## Key Takeaway

JOINs allow combining information from multiple tables.
Most real-world analysis requires joining data before aggregation.
