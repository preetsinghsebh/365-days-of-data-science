```python
import pandas as pd
import sqlite3

df = pd.read_csv("ecommerce_cleaned.csv")
conn = sqlite3.connect(":memory:")
df.to_sql("orders", conn, index=False, if_exists="replace")
```

```python
pd.read_sql(
    "SELECT SUM(total_amount) AS total_revenue FROM orders;",
    conn
)
```

```python
pd.read_sql(
    "SELECT product_category, SUM(total_amount) AS revenue FROM orders GROUP BY product_category;",
    conn
)
```

```python
pd.read_sql(
    "SELECT customer_id, SUM(total_amount) AS total_spent FROM orders GROUP BY customer_id ORDER BY total_spent DESC LIMIT 5;",
    conn
)
```

```python
pd.read_sql(
    "SELECT AVG(total_amount) AS avg_order_value FROM orders;",
    conn
)
```
