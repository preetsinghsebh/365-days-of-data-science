```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Car details v3.csv")
```

```python
avg_price_fuel = df.groupby("fuel")["selling_price"].mean()
avg_price_fuel = avg_price_fuel.sort_values(ascending=False)

avg_price_fuel
```

```python
avg_price_fuel.plot(kind="bar")
plt.title("Average Selling Price by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Average Price")
plt.show()
```

### Insight

Diesel cars generally have higher average selling prices
compared to petrol and other fuel types.

Fuel type influences resale value.
