```python
import pandas as pd

df = pd.read_csv("ecommerce_cleaned.csv")
df.head()
```

```python
avg_order_value = df.groupby("product_category")["total_amount"].mean()
avg_order_value
```

```python
avg_order_value = avg_order_value.sort_values(ascending=False)
avg_order_value
```

```python
category_orders = df.groupby("product_category")["order_id"].count()
category_revenue = df.groupby("product_category")["total_amount"].sum()

comparison = pd.DataFrame({
    "total_orders": category_orders,
    "total_revenue": category_revenue,
    "avg_order_value": avg_order_value
})

comparison.sort_values("avg_order_value", ascending=False)
```

### Day 41 Notes

Calculated average order value (AOV) per product category.
Compared order volume, total revenue, and AOV.

High revenue does not always mean high average order value.
