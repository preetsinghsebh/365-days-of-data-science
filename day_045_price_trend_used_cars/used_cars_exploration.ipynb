```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Car details v3.csv")
```

```python
df[["year", "selling_price"]].head()
```

```python
avg_price = df.groupby("year")["selling_price"].mean()
avg_price = avg_price.sort_index()

avg_price.head()
```

```python
avg_price.plot(kind="line")
plt.title("Average Selling Price by Year")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```

### Insight

Newer cars generally have higher average selling prices.
This shows depreciation effect over time.
